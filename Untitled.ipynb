```python
%matplotlib inline
import numpy as np
import theano
import theano.tensor as tt
import pymc3 as pm
import arviz
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

# 1. ranking based only on movie ranking

```python
dataframe = pd.read_csv("zestawy-danych/dataframe 0.csv")
dataframe.head()
```

```python
n_movies = len(dataframe.movie.unique())
movie_idx = dataframe.movie.values
print(n_movies)
```

```python
with pm.Model() as gmm:
   
    sigma_movie = pm.Gamma('sigma_movie', alpha=10, beta=5)
    mu_movie = pm.Normal('mu_movie', mu=0.0, sigma=sigma_movie, shape=n_movies)
    like_est = mu_movie[movie_idx]
    
    y = pm.Normal('y', sigma=sigma_movie,mu = like_est, observed=dataframe.like)
    trace = pm.sample(1000, tune=500)
    
    pm.traceplot(trace)
    summary = pm.summary(trace, var_names=['mu_movie'])
    summary.sort_values(by=['r_hat'])
    display(summary)

    pm.forestplot(trace, var_names=['mu_movie'])
    
```

# 2. Movie ranking based on movie,person and genre ranking

```python
n_movies = len(dataframe.movie.unique())
movie_idx = dataframe.movie.values

n_person = len(dataframe.person.unique())
person_idx = dataframe.person.values

n_genre = len(dataframe.genre.unique())
genre_idx = dataframe.genre.values


with pm.Model() as gmm:
   
    sigma_movie = pm.Gamma('sigma_movie', alpha=10, beta=5)
    sigma_person = pm.Gamma('sigma_person', alpha=10, beta=5)
    sigma_genre = pm.Gamma('sigma_genre', alpha=10, beta=5)
    
    mu_movie = pm.Normal('mu_movie', mu=0.0, sigma=sigma_movie, shape=n_movies)
    mu_person = pm.Normal('mu_person', mu=0.0, sigma=sigma_person, shape=n_person)
    mu_genre = pm.Normal('mu_genre', mu=0.0, sigma=sigma_genre, shape=n_genre)
    
    like_est = mu_movie[movie_idx] + mu_person[person_idx] + mu_genre[genre_idx]
    
    y = pm.Normal('y', sigma=sigma_movie,mu = like_est, observed=dataframe.like)
    trace = pm.sample(1000, tune=500)
    
    pm.traceplot(trace)
    summary = pm.summary(trace, var_names=['mu_movie'])
    summary.sort_values(by=['r_hat'])
    display(summary)

    pm.forestplot(trace, var_names=['mu_movie'])
```

# 2. Movie ranking based on movie,person,genre ranking and personal favourite genre 

```python
with pm.Model() as gmm:
   
    sigma_movie = pm.Gamma('sigma_movie', alpha=10, beta=5)
    sigma_person = pm.Gamma('sigma_person', alpha=10, beta=5)
    sigma_genre = pm.Gamma('sigma_genre', alpha=10, beta=5)
    
    mu_movie = pm.Normal('mu_movie', mu=0.0, sigma=sigma_movie, shape=n_movies)
    mu_person = pm.Normal('mu_person', mu=0.0, sigma=sigma_person, shape=n_person)
    mu_genre = pm.Normal('mu_genre', mu=0.0, sigma=sigma_genre, shape=n_genre)
    mu_gp = pm.Normal('mu_gp', mu=0.0, sigma=sigma_person, shape=(n_genre, n_person))
    
    like_est = mu_movie[movie_idx] + mu_person[person_idx] + mu_genre[genre_idx] + mu_gp[genre_idx, person_idx]
    
    y = pm.Normal('y', sigma=sigma_movie,mu = like_est, observed=dataframe.like)
    trace = pm.sample(1000, tune=500)
    
    pm.traceplot(trace)
    summary = pm.summary(trace, var_names=['mu_movie'])
    summary.sort_values(by=['r_hat'])
    display(summary)

    pm.forestplot(trace, var_names=['mu_movie'])
```
